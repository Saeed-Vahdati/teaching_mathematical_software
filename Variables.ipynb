بدست آوردت ریشه‌های معادله درجه ۲ با استفاده از ضرایب آن

```python
a = 3
b = 7
c = -8
delta = b ** 2 - 4 * a * c
x_1 = (-b + delta ** (1 / 2)) / (2 * a)
x_2 = (-b - delta ** (1 / 2)) / (2 * a)
print('x_1=', x_1)
print('x_2=', x_2)
```
