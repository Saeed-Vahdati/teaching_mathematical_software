در این برنامه می خواهیم نمودار تابع
$\frac{\sin(x)}{x^2 + \cos(x)}$
را رسم کنیم


```python
import numpy as np
x = np.linspace(-2,2,100)
y = np.sin(x) / (x ** 2 + np.cos(x))
```

```python
import matplotlib.pyplot as plt
plt.plot(x,y)
plt.xlabel('axis X')
plt.ylabel('axis Y')
plt.grid()
plt.show()
```
